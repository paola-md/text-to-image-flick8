```python
import pandas as pd
import numpy as np
import re
import string
import nltk
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split

from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer

# GLOBALS
TOKEN_DIR = "./Flickr8k_text/"
```

```python
from pickle import load, dump
with open("./pickles/PACO/token_train_64.pickle","rb") as f:
  synops_train = load(f)

with open("./pickles/PACO/token_test_64.pickle","rb") as f:
  synops_test = load(f)

with open("./pickles/PACO/token_val_64.pickle","rb") as f:
  synops_val = load(f)
```

```python
print(len(synops_train))
print(len(synops_test))
print(len(synops_val))
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('bert-large-nli-mean-tokens')
test_embeddings = model.encode(synops_test)
train_embeddings = model.encode(synops_train)
val_embeddings = model.encode(synops_val)
```

```python
from pickle import dump, load

with open("./pickles/PACO/E_train_LARGE.pickle","wb") as f:
  pickle.dump(train_embeddings, f)

with open("./pickles/PACO/E_test_LARGE.pickle","wb") as f:
  pickle.dump(test_embeddings, f)

with open("./pickles/PACO/E_val_LARGE.pickle","wb") as f:
  pickle.dump(val_embeddings, f)
```
